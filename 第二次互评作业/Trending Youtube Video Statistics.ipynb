## Find Frequent Patterns

```python
class Apriori:
    def __init__(self, datas,  minSupport=0.5, minConf=0.3):
        self.datas = datas
        self.minSupport = minSupport
        self.minConf = minConf
    
    @staticmethod
    def createC1(dataSet):
        C1 = []
        for transaction in np.array(dataSet):
            for item in transaction:
                if [item] not in C1:
                    C1.append( [item] )
        C1.sort()
        return list(map( frozenset, C1 ))
    
    @staticmethod
    def scanD( D, Ck, minSupport ):
        ssCnt = {}
        for tid in D:
            if Ck is not None:
                for can in Ck:
                    if can.issubset( tid ):
                        ssCnt[ can ] = ssCnt.get( can, 0) + 1
        numItems = float( len( D ) )
        retList = []
        supportData = {}
        for key in ssCnt:
            support = ssCnt[ key ] / numItems
            if support >= minSupport:
                retList.insert( 0, key )
                supportData[ key ] = support
        return retList, supportData
    
    @staticmethod
    def aprioriGen( Lk, k ):
        retList = []
        lenLk = len( Lk )
        for i in range( lenLk ):
            for j in range( i + 1, lenLk ):
                L1 = list( Lk[ i ] )[ : k - 2 ];
                L2 = list( Lk[ j ] )[ : k - 2 ];
                L1.sort();L2.sort()    
                if L1==L2:
                    retList.append( Lk[ i ] | Lk[ j ] ) 
        return retList
    
    def calcConf(self, freqSet, H, supportData, brl):
        prunedH = []
        lift = []
        file = open("generate_rules.txt","a",encoding = "utf-8")
        for conseq in H:
            conf = supportData[freqSet] / supportData[freqSet - conseq]
            if conf >= self.minConf:
                file.write(str(freqSet - conseq)+"-->"+str(conseq)+" support:"+str(supportData[freqSet])+" conf:"+str(conf)+'\n')
                brl.append((freqSet - conseq, conseq, supportData[freqSet], conf))
                prunedH.append(conseq)
        file.close()
        return prunedH


    def rulesFromConseq(self, freqSet, H, supportData, brl):
        m = len(H[0])
        if (len(freqSet) > m + 1):
            Hmp1 = self.aprioriGen(H, m + 1)
            Hmp1 = calcConf(freqSet, Hmp1, supportData, brl, self.minConf)
            if (len(Hmp1) > 1):
                self.rulesFromConseq(freqSet, Hmp1, supportData, brl, self.minConf)

    
    def generateRules(self, L, supportData):
        bigRuleList = []
        for i in range(1, len(L)):
            for freqSet in L[i]:
                H1 = [frozenset([item]) for item in freqSet]
                if (i > 1):
                    self.rulesFromConseq(freqSet, H1, supportData, bigRuleList)
                else:
                    self.calcConf(freqSet, H1, supportData, bigRuleList)
        return bigRuleList


    def lift_eval(self, rules, suppData):
        lift = []
        for rule in rules:
            freqSet_conseq = rule[0]
            conseq = rule[1]
            lift_val = float(rule[3]) / float(suppData[rule[1]])
            lift.append([freqSet_conseq,conseq,lift_val])
        return lift
    
    
    def __call__(self):
        C1 = self.createC1(self.datas )
        D =list( map( set, self.datas ))
        L1, suppData = self.scanD( D, C1, self.minSupport )
        L = [ L1 ]
        k = 2

        while ( len( L[ k - 2 ] ) > 0 ):
            Ck = self.aprioriGen( L[ k - 2 ], k )
            Lk, supK = self.scanD( D, Ck, self.minSupport )
            suppData.update( supK )
            L.append( Lk )
            k += 1
        return L, suppData
```

```python
my_df = pd.get_dummies(my_df)
apri = Apriori(datas=list(map(set,np.array(my_df))), minSupport=0.3, minConf=0.5)
L, suppData = apri()
print(L)
rules = apri.generateRules(L, suppData)
print(rules)
lifts = apri.lift_eval(rules, suppData)
print(lifts)
```
